Breast cancer detection using Logistic regression

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
breast_cancer=load_breast_cancer()
dir(breast_cancer)
```

```python
#using dataset present in sklearn 
import pandas as pd
df=pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
df.head()
```

```python
df.shape
df['class']=breast_cancer.target
df.head()
```

```python
x=df.drop('class',axis='columns')
y=breast_cancer.target
#df.groupby('class').mean()
```

```python
#splitting data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, stratify=y)
#test_size--->to split the data in 
```

Building Logistic Regression model 

```python
from sklearn.linear_model import LogisticRegression
lreg=LogisticRegression(max_iter=150)
lreg.fit(x_train,y_train)
print(lreg.score(x_train,y_train))
print(lreg.score(x_test,y_test))
```

```python
#detecting
input_data=[[17.99,10.38	,122.80,1001.0,0.11840,0.27760,0.3001,0.14710,0.2419,0.07871,1.0950,0.9053,8.589,153.40,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.60,2019.0,0.1622,0.6656,0.7119,0.2654,0.4601,0.11890]]
predicted_y=lreg.predict(input_data)
if(predicted_y[0]==0):
  print('This breast cancer is Malignant')
else:
  print('This breast cancer is Benign')
```
